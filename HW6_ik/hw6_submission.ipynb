```python
# 1 ===========================================================
import sys
sys.path.append("/home/daniel/software/MEEN537/")
import kinematics as kin
from visualization import VizScene
import sympy as sp
import numpy as np
import time
np.set_printoptions(precision=4, suppress=True)


def getTau(q, F, arm):
    J = arm.jacob(q)[:3, :] # Only need the linear velocity components of the Jacobian
    print(f"\n Tau for q={q} is:")
    sp.pprint(J.T @ F)

# Define the arm
a1 = a2 = 1
dh = [[0, 0, a1, -np.pi/2.0],
      [0, 0, a2, 0]]
arm = kin.SerialArm(dh)


# Define different configurations and get tau
qa = [0, np.pi/4]
Fa = np.array([[-1], [0], [0]])
getTau(qa, Fa, arm)

qb = [0, np.pi/2]
Fb = np.array([[-1], [0], [0]])
getTau(qb, Fb, arm)

qc = [np.pi/4, np.pi/4]
Fc = np.array([[-1], [-1], [0]])
getTau(qc, Fc, arm)

qd = [0, 0]
Fd = np.array([[0], [0], [1]])
getTau(qd, Fd, arm)

qe = [0, 0]
Fe = np.array([[1], [0], [0]])
getTau(qe, Fe, arm)
```
